### Example: Quadratic polynomial

```python
poly1 = decompose_polynomial('x^2+5*x-3')
p1 = Polynomial(poly1)
p1.__dict__
```

```python
# evaluate at some value, say x = 3
p1(3)
```

```python
# print and plot
print(p1)
p1.plot(x_min=-10, x_max=5)
```

```python
# find roots from plot
p1.newton(0.0)
```

```python
p1.newton(-6.0)
```

### Some addition and multiplication operations

```python
poly2 = decompose_polynomial("5*x^3+3*x^2+x-6")
poly3 = decompose_polynomial("7*x^8+3*x^5+2*x^3+2*x-9")
```

Create instance of `poly2`

```python
p2 = Polynomial(poly2)
p2.__dict__
```

```python
print(f"p2: {p2}")
print(f"Derivative: {p2.derivative()}")
```

Create instance of `poly3`

```python
p3 = Polynomial(poly3)
p3.__dict__
```

```python
print(f"p3: {p3}")
print(f"Derivative: {p3.derivative()}")
```

### Addition

```python
p2_p3 = p2 + p3
print(p2_p3)
```

### Multiplication

```python
p2p3 = p2 * p3
print(p2p3)
```
